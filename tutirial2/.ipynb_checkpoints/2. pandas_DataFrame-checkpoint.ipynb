1. Series가 index-value 매칭이었다면
2. Dataframe 은 표, 또는 행렬 형태로써 row, column에 대해 value가 맵핑

```python
import pandas as pd
```

```python

data = {                         # dict형 자료형을 DataFrame형으로 바꿀 수 있습니다.
    'id' : [1,2,3],
    'name' : ['Kim', 'Lee', 'Park'],
    'age' : [10,20,30],
    'asset' : [15.2, 11.2, 32.9],
    'job' : ['Farmer', 'Singer', 'CEO']
}
df = pd.DataFrame(data)
df                              # Series와 달리 index, column, value형태로 이루어진 것을 알 수 있습니다
                                # Column은 열의 제목, index는 행의 제목 입니다.
    
```

```python

# Columns과 index도 지정해 줄 수 있습니다. columns와 data값이 일치하지 않으면 남는 cell은 NaN으로 채워집니다.

df2 = pd.DataFrame(data,
                   columns=['id','name','age','asset','job','persona'],
                   index=['one','two','three']
                  ) 
df2
```

```python

df2['id']   # Column값으로 index와 value들을 출력할 수 있습니다.
```

```python

df2.loc['two'] # index값으로 출력하려면 .loc()를 써서 column과 value값들을 출력할 수 있다.
```

```python

df2['hobby'] = 'codiing' # 모든 hobby Column에 값을 넣는다. 만약에 hobby Column이 없다면, 새로만든다.
df2
```

```python

df2['age'] = [21,23,34] # age Column에 차례대로 21,23,34를 넣는다. numpy array로도 가능하다
df2
```

```python

# DataFrame에 Series도 넣을 수 있다. 그러면 Series의 해당 index에 value가 원하는 Column에 들어가게 된다.
series_temp = pd.Series( ['서울','경산','강릉'], index=['one','two','three'] ) 
df2['residence'] = series_temp
df2
```

```python

del df2['persona']         # 해당 Column을 제거합니다.
df2
```

```python

df2.index = df2['id'] # id Column값을 index로 만들기. one -> 1
df2
```

```python

df2.loc[4] = [5,'Nam',23, 20.2, 'student', 'club', '부산']
df2 = df2[ df2.id != 3 ]#df2.drop('4') # row도 제거가능, drop써도됨
df2
```

```python

df2.T # 전치
```

```python

df2['id'].values # Column 값 array로 추출
```

```python

df2.loc[1].values # row 1의 값 array로 추출
```

```python

df2.loc[1]['name'] # 특정 row, column 값 추출

# 특정 셀 값바꾸기 df2.loc[1]['name'] = "ㅁㄴㅇㄴㅇ" 는 안됨
df2.loc[1,'name'] = '이름' #으로 해야함
```


## 실습


1. 구구단 1단부터 3단까지 출력하는 데이터프레임 df1을 생성하라(컬럼3x로우9).

```python

df = pd.DataFrame(columns=range(1,4,1), index=range(1,10,1))
for i in range(1,4,1):
    for j in range(1,10,1):
        df.loc[j][i] = i*j
df
```


2. 인덱스를 숫자에서 한글로 바꿔라.
1 = '일', 2='이'....


```python
index_tmp = ['일','이','삼','사','오','육','칠','팔','구']
df.index = index_tmp
df
```


3. 4단을 추가하라


```python
df[4] = df[1]*4
df
```


4. 4단을 삭제하라


```python
del df[4]
df
```

5. 인덱스 '오', '칠' 을 제거하라

```python
df=df.drop(['오','칠']) # 대입해주자
df
```
